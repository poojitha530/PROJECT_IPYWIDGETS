```python
import ipywidgets as w
import math

h = w.Text()  # Input Box
output = w.FloatText()  # Output Box

# First row
buttonRad = w.Button(description="Rad")
buttonDeg = w.Button(description="Deg")
buttonFact = w.Button(description="x!")
buttonopen = w.Button(description="(")
buttonclose = w.Button(description=")")
buttonmod = w.Button(description="%")
buttonAC = w.Button(description="AC")

# Second row of Boxes
buttonInv = w.Button(description="Inv")
buttonSin = w.Button(description="Sin")
buttonLn = w.Button(description="Ln")
button7 = w.Button(description="7")
button8 = w.Button(description="8")
button9 = w.Button(description="9")
buttondiv = w.Button(description="/")

# Third row of Boxes
buttonPie = w.Button(description="π")
buttonCos = w.Button(description="Cos")
buttonLog = w.Button(description="Log")
button4 = w.Button(description="4")
button5 = w.Button(description="5")
button6 = w.Button(description="6")
buttonmul = w.Button(description="x")

# Fourth row of Boxes
buttone = w.Button(description="e")
buttonTan = w.Button(description="Tan")
buttonroot = w.Button(description="√")
button1 = w.Button(description="1")
button2 = w.Button(description="2")
button3 = w.Button(description="3")
buttonminus = w.Button(description="-")

# Fifth row of Boxes
buttonAns = w.Button(description="Ans")
button10power = w.Button(description="10^x")
buttonpower = w.Button(description="x^y")
button0 = w.Button(description="0")
buttondot = w.Button(description=".")
buttonequal = w.Button(description="=")
buttonequal.style.button_color = "ChartReuse"
buttonplus = w.Button(description="+")

# Backspace Button
buttonBS = w.Button(description="BackSpace")
buttonBS.style.button_color = "Orange"

first_box = w.HBox([buttonRad, buttonDeg, buttonFact, buttonopen, buttonclose, buttonmod, buttonAC])
second_box = w.HBox([buttonInv, buttonSin, buttonLn, button7, button8, button9, buttondiv])
third_box = w.HBox([buttonPie, buttonCos, buttonLog, button4, button5, button6, buttonmul])
fourth_box = w.HBox([buttone, buttonTan, buttonroot, button1, button2, button3, buttonminus])
fifth_box = w.HBox([buttonAns, button10power, buttonpower, button0, buttondot, buttonequal, buttonplus])

out_box = w.HBox([output, buttonBS])
box = w.VBox([h, first_box, second_box, third_box, fourth_box, fifth_box, out_box])

tab = w.Tab([box])
tab.set_title(0, "Calculator")
display(tab)

# Functions of 1st row Buttons
def fRad(self):
    h.value = h.value + "radians("

def fDeg(self):
    h.value = h.value + "degrees("

def fFact(self):
    h.value = h.value + "factorial("

def fopen(self):
    h.value = h.value + "("

def fclose(self):
    h.value = h.value + ")"

def fmod(self):
    h.value = h.value + "%"

def fAC(self):
    h.value = ""
    output.value = ""

# Functions of 2nd row buttons
def fInv(self):
    h.value = h.value + "1/("

def fSin(self):
    h.value = h.value + "math.sin("  # Update to use math.sin

def ln(self):
    h.value = h.value + "math.log("  # Update to use math.log

def f7(self):
    h.value = h.value + "7"

def f8(self):
    h.value = h.value + "8"

def f9(self):
    h.value = h.value + "9"

def fdiv(self):
    h.value = h.value + "/"

# Functions of 3rd row buttons
def fpie(self):
    h.value = h.value + "(22/7)"

def fCos(self):
    h.value = h.value + "math.cos("  # Update to use math.cos

def f4(self):
    h.value = h.value + "4"

def f5(self):
    h.value = h.value + "5"

def f6(self):
    h.value = h.value + "6"

def fmul(self):
    h.value = h.value + "*"

# Functions of 4th row buttons
def fe(self):
    h.value = h.value + ("2.71828**")

def fTan(self):
    h.value = h.value + "math.tan("  # Update to use math.tan

def square_root(self):
    h.value = h.value + "math.sqrt("  # Update to use math.sqrt

def f1(self):
    h.value = h.value + "1"

def f2(self):
    h.value = h.value + "2"

def f3(self):
    h.value = h.value + "3"

def fminus(self):
    h.value = h.value + "-"

# Functions of 5th row buttons
def f10power(self):
    h.value = h.value + "(10**"

def fpower(self):
    h.value = h.value + "**"

def f0(self):
    h.value = h.value + "0"

def fdot(self):
    h.value = h.value + "."

def fequal(self):
    s = eval(h.value)  # Use eval to calculate the expression
    output.value = s

def fplus(self):
    h.value = h.value + "+"

def fbackspace(self):
    h.value = h.value[0:-1]

# Row 1 function calling by onclick
buttonRad.on_click(fRad)
buttonDeg.on_click(fDeg)
buttonFact.on_click(fFact)
buttonopen.on_click(fopen)
buttonclose.on_click(fclose)
buttonmod.on_click(fmod)
buttonAC.on_click(fAC)

# Row 2 function calling by onclick
buttonInv.on_click(fInv)
buttonSin.on_click(fSin)
buttonLn.on_click(ln)
button7.on_click(f7)
button8.on_click(f8)
button9.on_click(f9)
buttondiv.on_click(fdiv)

# Row 3 function calling by onclick
buttonPie.on_click(fpie)
buttonCos.on_click(fCos)
buttonLog.on_click(ln)
button4.on_click(f4)
button5.on_click(f5)
button6.on_click(f6)
buttonmul.on_click(fmul)

# Row 4 function calling by onclick
buttone.on_click(fe)
buttonTan.on_click(fTan)
buttonroot.on_click(square_root)
button1.on_click(f1)
button2.on_click(f2)
button3.on_click(f3)
buttonminus.on_click(fminus)

# Row 5 function calling by onclick
buttonAns.on_click(fequal)
button10power.on_click(f10power)
buttonpower.on_click(fpower)
button0.on_click(f0)
buttondot.on_click(fdot)
buttonequal.on_click(fequal)
buttonplus.on_click(fplus)
buttonBS.on_click(fbackspace)
```
